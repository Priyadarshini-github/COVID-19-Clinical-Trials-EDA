```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/CUB/Downloads/COVID clinical trials.csv')
```

```python
print("First 5 rows of the dataset:")
print(df.head())
```

```python
print("Last 5 rows of the dataset:")
print(df.tail())
```

```python
print("\nDataFrame Info:")
print(df.info())
```

```python
print("\nSummary Statistics for Numerical Columns:")
print(df.describe())
```

```python
print("\nSummary Statistics for Categorical Columns:")
print(df.describe(include='object'))
```

```python
print("\nMissing Values Count:")
print(df.isnull().sum())
```

```python
df = df.drop(columns=['Acronym', 'Study Documents'])
```

```python
df.shape
```

```python
print("\nStatus Distribution: ")
print(df['Status'].value_counts())
df['Status'].value_counts().plot(kind='bar', title='Status of Clinical Trials')
plt.xlabel('Status')
plt.ylabel('Number of Trials')
plt.show()
```

```python
print("\nPhase Distribution:")
print(df['Phases'].value_counts())
df['Phases'].value_counts().plot(kind='bar', title='Distribution of Phases')
plt.xlabel('Phase')
plt.ylabel('Number of Trials')
plt.show()
```

```python
df['Results First Posted'].fillna('Unknown', inplace=True)
```

```python
age_counts = df['Age'].value_counts()
threshold = 0.02  # Example threshold (2%)
age_counts = age_counts[age_counts / age_counts.sum() > threshold]
age_counts = age_counts.append(pd.Series(df['Age'].value_counts()[df['Age'].value_counts() / df['Age'].value_counts().sum() <= threshold].sum(), index=['Other']))
plt.figure(figsize=(10, 4))
age_counts.plot(kind='bar')
plt.title('Age Group Distribution (Aggregated)')
plt.xlabel('Age Group')
plt.ylabel('Number of Trials')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Study Type', order=df['Study Type'].value_counts().index)
plt.title('Distribution of Study Types')
plt.xticks(rotation=90)
plt.xlabel('Study Type')
plt.ylabel('Number of Trials')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Gender', order=df['Gender'].value_counts().index)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Trials')
plt.show()
```

```python
status_phase = pd.crosstab(df['Status'], df['Phases'])
print("\nStatus vs. Phases Crosstab:")
print(status_phase)
status_phase.plot(kind='bar', stacked=True, title='Status vs. Phases')
plt.xlabel('Status')
plt.ylabel('Number of Trials')
plt.show()
```

```python
conditions_outcomes = df.groupby('Conditions')['Outcome Measures'].apply(lambda x: ', '.join(x.dropna().astype(str))).reset_index()

print("\nConditions and Outcome Measures:")
print(conditions_outcomes)
```

```python
numerical_cols = df.select_dtypes(include=[pd.np.number]).columns
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Apply PCA to reduce dimensions
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df[numerical_cols].dropna())
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2')
plt.title('PCA Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Phases', y='Enrollment')
plt.title('Scatter Plot of Enrollment by Phases')
plt.xlabel('Phases')
plt.ylabel('Enrollment')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
df['Primary Completion Date'] = pd.to_datetime(df['Primary Completion Date'], errors='coerce')
```

```python
trials_started_over_time = df['Start Date'].dt.to_period('M').value_counts().sort_index()
plt.figure(figsize=(12, 6))
trials_started_over_time.plot(kind='line', marker='o')
plt.title('Trials Started Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Trials Started')
plt.grid(True)
plt.show()
```

```python
trials_completed_over_time = df['Primary Completion Date'].dt.to_period('M').value_counts().sort_index()
plt.figure(figsize=(12, 6))
trials_completed_over_time.plot(kind='line', marker='o', color='orange')
plt.title('Trials Completed Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Trials Completed')
plt.grid(True)
plt.show()
```

```python
mean_enrollment = df.groupby('Phases')['Enrollment'].mean().reset_index()
heatmap_data = mean_enrollment.pivot('Phases', 'Enrollment', 'Enrollment')
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Heatmap of Mean Enrollment by Phases')
plt.xlabel('Enrollment')
plt.ylabel('Phases')
plt.show()
```
